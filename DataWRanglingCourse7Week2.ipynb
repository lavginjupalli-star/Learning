#Data Imputation
Data imputation is the process of replacing missing or incomplete data with estimated or substituted values. This helps maintain data integrity and ensures that analysis or machine learning models are not biased due to missing values.

When you have a dataset, sometimes some values are missing (blank or NaN).

Data imputation means filling those missing values with reasonable estimates so your analysis or machine learning model isn’t messed up.

Without imputation, models might give wrong results, because missing values create bias or errors.

#Handling Missing Values:


#1. Numerical Columns

Examples: Salary, Age, Price, Height

Ways to handle missing numbers:


--Mean (average) → good for normally distributed data.

--Median (middle value) → better when data has outliers.

--Mode (most frequent number) → for categorical-like numbers.

--Interpolation (Forward/Backward Fill)
   Guess the missing value based on nearby values.

   Example: If age values are [20, NaN, 22], you can fill the missing with 21.

--Removing Rows
If too many values are missing in a row (e.g., 70% empty), sometimes it’s best to delete the row.

#2. Categorical Columns

Examples: Gender, City, Product Category


--Fill with Mode (most frequent value)
If “City” is missing, and “Toronto” appears most often, fill with “Toronto.”

--Fill with 'Unknown' or 'Other'
Helps when you don’t want to assume anything.
Example: If “Gender” is missing, replace with “Unknown.”

--One-Hot Encoding with NaN Handling
When converting categories to numbers (dummy variables), treat missing values as a separate category.
Example: “City = Toronto, New York, NaN” → becomes three columns: Toronto (1/0), New York (1/0), Missing (1/0).

👉 In short:

Numbers → use mean, median, interpolation, or drop.

Categories → use mode, “Unknown,” or encode them properly.

#**DATA NORMALIZATION**

**Why do we normalize data before modeling?**
Data normalization = transforming data values into a common scale without distorting differences in ranges.

**The main purpose: make data consistent, comparable, and better for analysis or machine learning models.**

Bring values to same scale

Prevent bias toward large values

Improve ML performance

Data normalization is the process of scaling numerical values to a standard range, often [0, 1] or [-1, 1]. This is a critical step in data wrangling, especially when dealing with machine learning models, as normalization ensures that each feature contributes equally.

There are three common techniques for normalization:

#why do we normalize?



Example:

Age → range 0–100

Salary → range 30,000–100,000

A model (like KNN, Logistic Regression, Neural Nets) might think Salary is more important just because the numbers are bigger.It happens because many models rely on distance, magnitude, or weights — and larger numbers dominate mathematically.

Models aren’t intelligent the way humans are. They don’t understand meaning, units, or context. They just follow math.

They see numbers, not concepts
They apply formulas (distance, weights, gradients)
Bigger numbers → bigger impact in calculations

So it’s not that they’re dumb — they’re literal and mechanical.

Normalization rescales so both features contribute fairly.

#Simple Feature Scaling

Take each value of a feature.

Divide by the maximum value in that feature.

Result → values are between 0 and 1.

```python
import pandas as pd
import numpy as np


# Creating a sample DataFrame with different feature ranges
data = {
    'Feature1': [10, 20, 30, 40, 50],  # Small range
    'Feature2': [100, 200, 300, 400, 500],  # Larger range
    'Feature3': [5, 500, 1000, 1500, 2000]  # Highly varied range
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)


df_simple_scaled = df / df.max()

print("\nSimple Feature Scaled DataFrame (Values between 0 and 1):")
print(df_simple_scaled)
```

1.   .**Simple Feature Scaling**


   - Scales the data by dividing each value by the maximum absolute value of the feature.
   - Formula: X_scaled = X / X_max
   - Ensures all values are in the range [0,1].
   - Suitable for cases where only positive values are expected.

   **Pros:**
   - Easy to implement and computationally efficient.
   - Works well when all values are positive.

   **Cons:**
   - Does not handle negative values well.
   - Sensitive to outliers as it depends on the max value.

#Min -Max Scaling

```python
# Min-Max Scaling without using Scaler
df_min_max = (df - df.min()) / (df.max() - df.min())

print("\nMin-Max Scaled DataFrame (Values between 0 and 1):")
print(df_min_max)
```



2. **Min-Max Scaling** (Normalization)
   - Scales the data to a fixed range, usually [0,1] or [-1,1].
   - Formula: X_scaled = (X - X_min) / (X_max - X_min)
   - Useful when we need to preserve relationships in data and when data is bounded.
   - Commonly used in algorithms like k-NN, neural networks, etc.

   **Pros:**
   - Preserves the original distribution of data.
   - Keeps all values within a fixed range, preventing large feature dominance.

   **Cons:**
   - Sensitive to outliers since min and max values are used for scaling.

#Z score

The Z-score normalization (also called standardization) is a method where the mean of the data is subtracted and then divided by the standard deviation. This transforms the data to have a mean of 0 and a standard deviation of 1. In this case, every feature will be scaled relative to its distribution.

You can do this directly in Python using the scipy.stats.zscore() function, or even without any external library using basic operations in pandas.

| Feature  | Normalization | Standardization          |
| -------- | ------------- | ------------------------ |
| Range    | 0 to 1        | No fixed range           |
| Uses     | Min & Max     | Mean & Std               |
| Outliers | Sensitive     | More robust              |
| Shape    | Changes scale | Keeps distribution shape |


**Simple analogy**

Normalization = “fit everything into same box”

Standardization = “adjust based on average behavior”

```python
# Z-score Standardization without using Scaler
df_zscore = (df - df.mean()) / df.std()

print("\nZ-score Standardized DataFrame (Mean = 0, Std Dev = 1):")
print(df_zscore)
```



3. **Z-score Standardization** (Standardization)
   - Transforms data to have a mean (μ) of 0 and a standard deviation (σ) of 1.
   - Formula: X_standardized = (X - μ) / σ
   - Useful when data follows a Gaussian (Normal) distribution.
   - Commonly used in algorithms like logistic regression, SVM, and PCA.

   **Pros:**
   - Less sensitive to outliers compared to Min-Max scaling.
   - Useful when features have different scales and units.

   **Cons:**
   - Does not bound data within a specific range, which might not be ideal for some ML models.

```python
from scipy.stats import zscore
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}

df = pd.DataFrame(data)

# Apply Z-score normalization
df[['Age', 'Salary']] = df[['Age', 'Salary']].apply(zscore)

print(df)
```
