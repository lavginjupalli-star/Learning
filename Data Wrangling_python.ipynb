**Creating Equal-Width Bins for Horsepower using NumPy**

To divide the "horsepower" column into 3 equal-width bins, we use NumPy’s linspace() function to generate evenly spaced boundary values.

🔹 Step 1: Define the range

We start by selecting the full range of the data:

The starting value is the minimum horsepower in the dataset, ensuring the smallest value is included.
The ending value is the maximum horsepower in the dataset, ensuring the largest value is included.

This guarantees that all data points fall within the defined range.

🔹 Step 2: Determine the number of boundaries

To create 3 bins, we need 4 boundary points, because the number of boundaries is always one more than the number of bins.

🔹 Step 3: Generate equal intervals

Using a function that creates evenly spaced values, we generate 4 points between the minimum and maximum horsepower values. These points act as the cut-off boundaries that split the data into equal sections.

🔹 Step 4: Formation of bins

The resulting boundaries divide the full range into three equal-width intervals:

First bin: from minimum value to first boundary
Second bin: from first boundary to second boundary
Third bin: from second boundary to maximum value

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Plot the distribution of each bin:
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

**Indicator Variable**

What is an indicator variable?

An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning.

Why use indicator variables?

You use indicator variables so you can use categorical variables for regression analysis.

Example
The column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, you can convert "fuel-type" to indicator variables.

Use the Panda method 'get_dummies' to assign numerical values to different categories of fuel type.

```python
df.head()
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
# merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)
# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```
